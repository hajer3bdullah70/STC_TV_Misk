```python
movie_features_df=dataframe.pivot_table(index='program_name',columns='user_id_maped',values='rating').fillna(0)
movie_features_df.head()
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
movie_features_df_matrix = csr_matrix(movie_features_df.values)
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_df_matrix)
```

```python
program_name = 'Moana'
recommendations = 5

# Find the index for the desired program
program_index = movie_features_df.index.get_loc(program_name)

# Get distances and matrices of nearest neighbors
distances, indices = model_knn.kneighbors(movie_features_df.iloc[program_index, :].values.reshape(1, -1), n_neighbors=recommendations+1)

# Print Recommednations
print(f'Recommendations for "{program_name}":\n')
for i in range(1, len(distances.flatten())):
    recommended_movie_index = indices.flatten()[i]
    distance = distances.flatten()[i]
    recommended_movie_name = movie_features_df.index[recommended_movie_index]
    print(f'{i}: {recommended_movie_name}, with a distance of {distance:.4f}')
```
